```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv("https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv")
df1
```

```python
df1[df1['Age']!="00 to 17"]
```

```python
def string(title):
    if '11' in title:
        return True
    else:
        return False
df1 = df1[df1['Year'].apply(lambda x: string(x))]
df1 = df1[df1['Age']!='00 to 17']
```

```python
df1
```

```python
df1.groupby("Educational Attainment")['Population Count'].sum()
```

```python
plt.pie(df1.groupby("Educational Attainment")['Population Count'].sum())
```

```python
contingency_table=pd.pivot_table(df1,index='Educational Attainment',columns='Personal Income',aggfunc='sum',fill_value=0)
contingency_table
```
